```python
def text_analysis(text):
    sentences = text.split('.')
    num_sentences = len(sentences)
    word_counts = [len(sentence.split()) for sentence in sentences]
    total_characters = len(text)
    word_counts += [0] * (7 - len(word_counts))

    return num_sentences, word_counts[0], word_counts[1], word_counts[2], word_counts[3], word_counts[4], word_counts[5], word_counts[6], total_characters

df['num_sentences'], df['words_in_1st_sentence'], df['words_in_2nd_sentence'], df['words_in_3rd_sentence'], df['words_in_4th_sentence'], df['words_in_5th_sentence'], df['words_in_6th_sentence'], df['words_in_7th_sentence'], df['total_characters'] = zip(*df['text'].apply(text_analysis))

df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Define representative phrases for each category
category_phrases = {
    1: "People anecdotes legends",
    2: "Philosophy religion history customs geography",
    3: "Environment resources recycling",
    4: "Physics chemistry biology earth science",
    5: "Sports leisure hobbies travel",
    6: "Music art movies dance photography architecture",
    7: "Education schools career",
    8: "Language literature culture",
    9: "Computers internet information media transportation",
    10: "Psychology interpersonal relationships",
    11: "Politics economy society law",
    12: "Medicine health nutrition food"
}

# Combine category phrases into a single list for vectorization
categories_combined = list(category_phrases.values())

# Initialize the TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Vectorize the category phrases
tfidf_category = tfidf_vectorizer.fit_transform(categories_combined)

# Function to classify text into the most similar category
def classify_by_similarity(text):
    if not isinstance(text, str):
        return float('nan')

    # Vectorize the input text
    tfidf_text = tfidf_vectorizer.transform([text])

    # Calculate cosine similarity between the text and each category
    cosine_similarities = cosine_similarity(tfidf_text, tfidf_category).flatten()

    # Find the category with the highest similarity
    most_similar_category = np.argmax(cosine_similarities) + 1  # Adding 1 to match category numbering

    # Return the most similar category
    return most_similar_category if cosine_similarities[most_similar_category - 1] > 0 else float('nan')

# Apply the classification function to the first 25 rows of the dataset
df['topic'] = df['text'].iloc[:].apply(classify_by_similarity)

# Display the updated dataframe with the first 25 rows
df.head(25)
```

```python
import stanza
from collections import defaultdict
stanza.download('en')
nlp = stanza.Pipeline(lang='en', processors='tokenize,mwt,pos,lemma')
def lemmatize_text(text):
    doc = nlp(text)
    lemmas = [word.lemma for sent in doc.sentences for word in sent.words]
    filtered_list = [word for word in lemmas if word.isalpha()]
    return filtered_list
```

```python
df_filtered['lemmas'] = df_filtered['text'].apply(lambda x: lemmatize_text(x) if isinstance(x, str) else '')
```
